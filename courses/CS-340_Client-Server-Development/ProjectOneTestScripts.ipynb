```python
import sys
sys.path.insert(0, '/home/codio/workspace/code_files')
from CRUD_Python_Module import AnimalShelter

# Instantiate - no parameters needed (credentials hardcoded in __init__)
shelter = AnimalShelter()
print("Connection established")
```

```python
# Create a new animal document
new_animal = {
    "name": "TestDog",
    "age_upon_outcome_years": 2.5,
    "animal_type": "Dog",
    "breed": "Labrador Mix",
    "outcome_type": "Adoption"
}

result = shelter.create(new_animal)
print(f"Create successful: {result}")
```

```python
# Create multiple animals at once
new_animals = [
    {
        "name": "TestCat",
        "age_upon_outcome_years": 1.0,
        "animal_type": "Cat",
        "breed": "Domestic Shorthair",
        "outcome_type": "Adoption"
    },
    {
        "name": "TestRabbit",
        "age_upon_outcome_years": 0.5,
        "animal_type": "Rabbit",
        "breed": "Holland Lop",
        "outcome_type": "Transfer"
    }
]

for animal in new_animals:
    result = shelter.create(animal)
    print(f"Inserted {animal['name']}: {result}")
```

```python
# Query for documents by breed
query = {"breed": "Labrador Retriever"}
results = shelter.read(query)

print(f"\nQuery: {query}")
print(f"Results count: {len(results)}")
if results:
    for doc in results[:3]:  # Print first 3 results
        print(f"  - {doc.get('name', 'Unknown')}: {doc.get('breed', 'Unknown')}")
```

```python
# Query for documents by outcome_type
query = {"outcome_type": "Transfer"}
results = shelter.read(query)

print(f"\nQuery: {query}")
print(f"Results count: {len(results)}")
if results:
    for doc in results[:3]:
        print(f"  - {doc.get('name', 'Unknown')}: {doc.get('outcome_type', 'Unknown')}")
```

```python
# Query with multiple criteria
query = {"breed": "Pit Bull", "outcome_type": "Transfer"}
results = shelter.read(query)

print(f"\nQuery: {query}")
print(f"Results count: {len(results)}")
if results:
    for doc in results[:3]:
        print(f"  - {doc.get('name', 'Unknown')}: Breed={doc.get('breed')}, Type={doc.get('outcome_type')}")
else:
    print("No documents found with those criteria")
```

```python
# Confirm total documents in animals collection
all_animals = shelter.read({})
print(f"\nTotal documents in animals collection: {len(all_animals)}")
```

```python
# INSERT TEST RECORD FOR UPDATE TESTING
test_update_animal = {
    "name": "UpdateTest",
    "age_upon_outcome_years": 1.0,
    "animal_type": "Dog",
    "breed": "Beagle",
    "outcome_type": "Adoption"
}

result = shelter.create(test_update_animal)
print(f"Test record created for update: {result}")
```

```python
# UPDATE TEST - Change a single field
query = {"name": "UpdateTest"}
update_data = {"outcome_type": "Transfer"}

modified = shelter.update(query, update_data)
print(f"Update result: {modified} document(s) modified")

# VERIFY THE UPDATE WORKED
updated_record = shelter.read({"name": "UpdateTest"})
if updated_record:
    print(f"Verified: {updated_record[0].get('name')} outcome is now {updated_record[0].get('outcome_type')}")
```

```python
# UPDATE TEST - Change multiple fields at once
query = {"name": "UpdateTest"}
update_data = {
    "age_upon_outcome_years": 2.5,
    "outcome_type": "Adoption"
}

modified = shelter.update(query, update_data)
print(f"Update result: {modified} document(s) modified")

# VERIFY THE UPDATES WORKED
updated_record = shelter.read({"name": "UpdateTest"})
if updated_record:
    print(f"Verified: Age={updated_record[0].get('age_upon_outcome_years')}, Outcome={updated_record[0].get('outcome_type')}")
```

```python
# INSERT TEST RECORD FOR DELETE TESTING
test_delete_animal = {
    "name": "DeleteTest",
    "age_upon_outcome_years": 0.5,
    "animal_type": "Cat",
    "breed": "Siamese",
    "outcome_type": "Adoption"
}

result = shelter.create(test_delete_animal)
print(f"Test record created for deletion: {result}")

# VERIFY IT EXISTS BEFORE WE DELETE IT
verify = shelter.read({"name": "DeleteTest"})
print(f"Record exists before delete: {len(verify) > 0}")
```

```python
# DELETE TEST - Remove a single document
query = {"name": "DeleteTest"}

deleted = shelter.delete(query)
print(f"Delete result: {deleted} document(s) deleted")

# VERIFY THE DELETION WORKED
verify = shelter.read({"name": "DeleteTest"})
print(f"Record exists after delete: {len(verify) > 0}")
```

```python
# DELETE TEST - Remove multiple documents at once
# Check how many records have "TestBreed" (should be 0 initially)
count_before = shelter.read({"breed": "TestBreed"})
print(f"Documents with TestBreed before delete: {len(count_before)}")

# Create a few test animals with "TestBreed"
for i in range(2):
    test_animal = {
        "name": f"DeleteMultiTest{i}",
        "breed": "TestBreed",
        "animal_type": "Dog",
        "outcome_type": "Transfer"
    }
    shelter.create(test_animal)

# Verify we created 2 records
count_created = shelter.read({"breed": "TestBreed"})
print(f"Documents with TestBreed after creating: {len(count_created)}")

# Delete ALL records with breed = "TestBreed"
deleted = shelter.delete({"breed": "TestBreed"})
print(f"Delete result: {deleted} document(s) deleted")

# Verify deletion - should be 0 now
verify = shelter.read({"breed": "TestBreed"})
print(f"Documents with TestBreed after delete: {len(verify)}")
```

```python
# FINAL VERIFICATION
all_animals = shelter.read({})
final_count = len(all_animals)
print(f"Final document count in animals collection: {final_count}")
print("All CRUD operations completed successfully!")
```
